 # AIRTEL Customer Churn EDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Read file

```python
df = pd.read_csv("Customer Churn.csv")
df
```

# INFORMATION OF DATA 

```python
df.info()
```

# REPLACE SPACES WITH 0 AND CHANGE TYPE OF TotalCharges

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype(float)
```

```python
df.info()
```

# CHECK NULL 

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

# CHECK DUPLICATES

```python
df["customerID"].duplicated().sum()
```

# converted 0 and 1 values of senior citizen to make it easier to understand

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
        
df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)        
```

```python
df.head()
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x = 'Churn',data=df)
ax.bar_label(ax.containers[0])
plt.title("Count Of Customers By Churn")
plt.show()
```

```python
plt.figure(figsize=(4,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'] , labels = gb.index , autopct = "%1.2f%%")
plt.title("Percentage Of Churned Customers")
plt.show()   
```

```python
plt.figure(figsize=(5,4)) 
ax = sns.countplot(x = 'gender', data = df , hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn By Gender")
plt.show
```

```python
plt.figure(figsize=(5,4)) 
ax = sns.countplot(x = 'SeniorCitizen', data = df )
ax.bar_label(ax.containers[0])
plt.title("Count Of Customers By SeniorCitizen")
plt.show 
```

```python
plt.figure(figsize=(10,5)) 
sns.histplot(x = "tenure", data = df , bins = 72 , hue = "Churn")
plt.title("Distribution of Customer Tenure by Churn Status")
plt.show()
```

```python
plt.figure(figsize=(8,4)) 
ax = sns.countplot(x = 'Contract', data = df , hue = 'Churn')
ax.bar_label(ax.containers[0])
plt.title("Count Of Customers By Contract")
plt.show 
```

# People Who Have Month To Month Contract Are Likely To Churn Then From Those Who Have 1 Or 2 Year Contact.

```python
df.columns
```

```python
cat_cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Define cat_cols (as above)
# ...

# Melt to long format
melted = df.melt(
    id_vars=["Churn"],                # or include columns you want to keep
    value_vars=cat_cols,
    var_name='feature',
    value_name='value'
)

# Plot with seaborn FacetGrid
sns.catplot(
    data=melted,
    x='value',
    col='feature',
    kind='count',
    hue = 'Churn',
    col_wrap=3,          # e.g. wrap 3 per row
    sharex=False,
    sharey=False,
    height=4,
    aspect=1
 
)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9,4)) 
ax = sns.countplot(x = 'PaymentMethod', data = df , hue = 'Churn')
ax.bar_label(ax.containers[0])


plt.title("Churn Of Customers By PaymentMethod")
plt.xticks(rotation = 45)
plt.show 
```
